## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

from {{ cookiecutter.project_module_name }}.utils.paths import RAW_DATA_DIR, FIGURES_DIR
from {{ cookiecutter.project_module_name }}.data.make_dataset import load_data

pd.set_option("display.max_columns", 50)
```

## 2. Cargar datos

Coloca el CSV en `data/raw/` y ajusta el nombre.

```python
DATA_FILE = "<nombre>.csv"   # ← ajusta
df = load_data(DATA_FILE)
df.head()
```

## 3. Inspección general

```python
print(df.shape)
df.info()
```

```python
df.describe(include="all").T
```

## 4. Valores nulos

```python
print(df.isnull().sum().sort_values(ascending=False))
msno.matrix(df)
plt.tight_layout()
plt.savefig(FIGURES_DIR / "missing_values.png", dpi=150)
plt.show()
```

## 5. Distribuciones

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()

df[num_cols].hist(bins=30, figsize=(14, 10), edgecolor="white")
plt.suptitle("Distribuciones numéricas")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "distributions.png", dpi=150)
plt.show()
```

## 6. Correlaciones

```python
corr = df[num_cols].corr()
fig, ax = plt.subplots(figsize=(max(8, len(num_cols)), max(6, len(num_cols)-1)))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, ax=ax)
plt.tight_layout()
plt.savefig(FIGURES_DIR / "correlation_matrix.png", dpi=150)
plt.show()
```
